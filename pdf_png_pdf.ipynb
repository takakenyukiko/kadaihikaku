```python
import os
from pathlib import Path
from io import BytesIO
from pdf2image import convert_from_path
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.lib.utils import ImageReader

from google.colab import files

# 複数ファイルをアップロード
uploaded = files.upload()

# 出力先フォルダ
output_dir = Path("/content/drive/MyDrive/Obsidian/pdfato")
output_dir.mkdir(parents=True, exist_ok=True)

def merge_pdf_pages_as_images(input_pdf, output_pdf):
    images = convert_from_path(str(input_pdf))

    width, height = letter
    total_height = height * len(images)

    packet = BytesIO()
    c = canvas.Canvas(packet, pagesize=(width, total_height))

    y_position = total_height

    for img in images:
        y_position -= height

        img_width, img_height = img.size
        scale = width / img_width
        scaled_height = img_height * scale

        img_io = BytesIO()
        img.save(img_io, format="PNG")
        img_io.seek(0)

        c.drawImage(
            ImageReader(img_io),
            0,
            y_position,
            width,
            scaled_height
        )

    c.save()
    packet.seek(0)

    with open(output_pdf, "wb") as f:
        f.write(packet.getvalue())

# アップロードされたPDFを順番に処理
for uploaded_filename in uploaded.keys():
    input_pdf = Path(uploaded_filename)

    if input_pdf.suffix.lower() != ".pdf":
        print(f"スキップ: {uploaded_filename} はPDFではありません")
        continue

    file_stem = input_pdf.stem
    output_pdf = output_dir / f"{file_stem}out.pdf"

    merge_pdf_pages_as_images(input_pdf, output_pdf)

    print(f"✅ PDF saved as {output_pdf}")
```

# 授業ノート用ドキュメントファイルからPNGへ変換し、/content/drive/MyDrive/Colab　Notebooks/pdfato/へ保存する

```python
import os
from pathlib import Path
from pdf2image import convert_from_path
from PIL import Image # PILライブラリを明示的にインポート（pdf2imageが依存）

# ...（ファイルのアップロードと名前の取得はそのまま）...

# ▼ ファイルアップロード用（ローカルファイル選択）
from google.colab import files
uploaded = files.upload()  # ユーザーがローカルからアップロード

# 最初にアップロードされたPDFファイルを取得
uploaded_filename = list(uploaded.keys())[0]
input_pdf = Path(uploaded_filename)

# 出力ファイル名（"〇〇_out.png"）
file_stem = os.path.splitext(os.path.basename(uploaded_filename))[0]
# 出力パスをGoogle Driveに設定
output_png = Path("/content/drive/MyDrive/Colab Notebooks/pdfato") / f"{file_stem}_merged.png"

def merge_pdf_pages_to_single_png(input_pdf: Path, output_png: Path):
    """
    PDFの全ページを縦に結合し、1枚のPNG画像として保存します。
    """
    print("⏳ Converting PDF pages to images...")
    # PDFのページをPIL Imageオブジェクトのリストに変換
    images = convert_from_path(input_pdf)

    if not images:
        print("🚨 Error: No pages found in the PDF.")
        return

    # 基準となる幅 (1ページ目の幅)
    base_width = images[0].width

    # 全ページの合計の高さと、リサイズ後の各画像の高さを計算
    # 全ての画像をbase_widthにリサイズしたときの高さを合計する
    total_height = 0
    resized_heights = []
    for img in images:
        # 縦横比を維持して、幅をbase_widthに合わせる
        scale = base_width / img.width
        scaled_height = int(img.height * scale)
        total_height += scaled_height
        resized_heights.append(scaled_height)

    # 結合用の新しい大きな画像（キャンバス）を作成
    # モードは'RGB' (カラー) または 'RGBA' (透明度が必要な場合)
    # PNGに保存するので'RGB'で十分です
    merged_image = Image.new('RGB', (base_width, total_height), color='white')

    # 各画像をキャンバスに貼り付ける
    current_y = 0
    for i, img in enumerate(images):
        # 1. 画像を結合後のサイズにリサイズ
        resized_img = img.resize((base_width, resized_heights[i]), Image.LANCZOS)

        # 2. 大きな画像に貼り付け
        # (左上隅の座標 (x, y) で指定)
        merged_image.paste(resized_img, (0, current_y))

        # 3. 次の画像の開始位置を更新
        current_y += resized_heights[i]

    # 結合された画像をPNG形式で保存
    merged_image.save(output_png, "PNG")
    print(f"\n✅ All pages merged and saved as single PNG: {output_png}")

# 関数を実行
merge_pdf_pages_to_single_png(input_pdf, output_png)
```

```python
import glob  # これを追加
import os
from pathlib import Path
from pypdf import PdfWriter
from pdf2image import convert_from_path
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter,A3
from io import BytesIO
from reportlab.lib.utils import ImageReader

def merge_pdf_pages_as_images(input_pdf, output_pdf):
    # PDFの各ページを画像として変換
    images = convert_from_path(input_pdf)

    # 各ページのサイズを取得（ここではLetterサイズに固定）
    width, height = letter
    total_height = height * len(images)

    # ReportLabでキャンバスを作成
    packet = BytesIO()
    c = canvas.Canvas(packet, pagesize=(width, total_height))

    # 各ページをキャンバスに描画
    y_position = total_height
    for img in images:
        y_position -= height
        img_width, img_height = img.size
        scale = width / img_width  # 横幅をページ幅に合わせてスケーリング
        scaled_height = img_height * scale

        # ImageReaderを使用してBytesIOを描画
        img_io = BytesIO()
        img.save(img_io, format="PNG")
        img_io.seek(0)
        c.drawImage(ImageReader(img_io), 0, y_position, width, scaled_height)

    # キャンバスを保存
    c.save()

    # 保存したキャンバスをPDFとして書き出し
    packet.seek(0)
    with open(output_pdf, "wb") as f:
        f.write(packet.getvalue())

# 入力ファイルと出力ファイルを指定
#input_pdf = "/content/drive/MyDrive/Colab Notebooks/pdfmoto/2年生4回目 - Google ドキュメント.pdf"
#output_pdf = "/content/drive/MyDrive/Colab Notebooks/pdfmoto/output.pdf"

files = glob.glob("/content/drive/MyDrive/Colab Notebooks/pdfmoto/*.pdf")
for file in files:
  print(file)

# 2. Googleアカウントにログイン
#from google.colab import auth
#auth.authenticate_user()

# 3. PDFファイルをColabにアップロード
#from google.colab import files
#uploaded = files.upload()
#files = list(uploaded.keys())[0]  # 最初のファイル名を取得


# poppler/binを環境変数PATHに追加する
#poppler_dir = Path( "/content/drive/MyDrive/Colab Notebooks/module")
#os.environ["PATH"] += os.pathsep + str(poppler_dir)

# PDFファイルのパス
input_pdf = Path(files[0])

file=os.path.splitext(os.path.basename(files[0]))[0]
output_pdf ="/content/drive/MyDrive/Colab Notebooks/pdfato/"+file+"out.pdf"

merge_pdf_pages_as_images(input_pdf, output_pdf)
print(f"PDF saved as {output_pdf}")
```
